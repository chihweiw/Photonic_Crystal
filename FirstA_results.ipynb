## Tasks 2.1 Use the numerical approach: 

We use MEEP to simulate an one dimensional system with an gaussian source and a film of $\epsilon_r=12$ and thickness $d=500nm$. The frequency range of the gaussian source is: $f_{min}=$ 300THz $< f <$ $f_{max}=$ 900THz. To Plot $T$ vs $\omega\big/\frac{\pi c}{n d}$, we compute the x-axis coordinate by $\omega\big/\frac{\pi c}{n d}=2\,\sqrt{\epsilon_r} d\,\frac{f}{c}$ for each frequency point. The numerical result (the red dots) match the analytical solution (the blue line):

$$T=\frac{1}{1+F\sin^2\frac{\delta}{2}}$$
$$F=\frac{4\,R}{(1-R)^2}$$
$$R= \big(\frac{\sqrt{\epsilon_r}-1}{\sqrt{\epsilon_r}+1}\big)^2$$
$$\delta = 2\,\sqrt{\epsilon_r} d\,\frac{\omega}{c} $$

Tasks 2.1: The blue line is the theoretical prediction and red dots are the results of the simulation.

## Tasks 2.3 Case (ABABAB...)

We now consider a structure with 15 repeated AB layers and Plot $T$ vs $\omega$ for such setup. From the following graph, we see there indeed a gap in the range from rougly $\omega=610$THz to $\omega=930$THz. According to the theory, the gap should be at $k\approx \pi/a$. So the corresponding wavelength $\lambda \approx 2\,a = 1\mu m$. According to the Weiner bounds, the effective dielectric constant $\epsilon_\alpha$ should be bounded around: $4.28<\epsilon_\alpha<6.6$. The corresponding frequency is bounded roughly by: $117$THz$<f<145$THz. This means the angular frequency $\omega$ should be around: $730$THz$<\omega<910$THz. This seems to agree the result of the simulation.

If we narrow the different between $\epsilon_A$ and $\epsilon_B$, the gap should become smaller. We find out that if we choose $\epsilon_A$=11 and $\epsilon_B$=4, the gap is indeed shrinking (now rougly 620THz~870THz).

For the next graph, we choose $\epsilon_A$=10 and $\epsilon_B$=5 and the gap become rougly 640THz~800THz.

For the next graph, we choose $\epsilon_A$=9 and $\epsilon_B$=6 and the gap become rougly 670THz~750THz. For this case, the effective dielectric constant is around $6.9<\epsilon_\alpha<7.2$. This means the center of the gap should be around $702$THz$<\omega<718$THz. This also agrees the result of the simulation.
